```

weight = spark.sql('select avg(case when Time_Deposits_Flag then 1 else 0 end) as perc_of_true from default.timedeposit_propensitymodeling_sample1_txt').head(1)[0][0]*1
```

```
df_t = spark.sql('select *, ''1'' as Flg from default.timedeposit_propensitymodeling_sample1_txt where Time_Deposits_Flag = true')
```

```
df_f = spark.sql('select *, ''0'' as Flg from default.timedeposit_propensitymodeling_sample1_txt where Time_Deposits_Flag = false').sample(False,weight)
```

```
df = df_t.union(df_f)
```

```
trainDF, testDF = df.drop('Birth_Date').drop('Ref_Date')\
  .drop('Customer_ID').drop('Time_Deposits_Flag').randomSplit([0.75, 0.25], seed=42)
```

```
nonNumeric = [x[0] for x in trainDF.dtypes if x[1]  not in ['int','tinyint','float','boolean']]
```

```
from pyspark.ml.feature import StringIndexer, OneHotEncoder
 

# The following two lines are estimators. They return functions that we will later apply to transform the dataset.
stringIndexer = StringIndexer(inputCols=nonNumeric, outputCols=[x + "Index" for x in nonNumeric]) .setHandleInvalid("keep") 
encoder = OneHotEncoder(inputCols=stringIndexer.getOutputCols(), outputCols=[x + "OHE" for x in nonNumeric]) 
 
# The label column ("income") is also a string value - it has two possible values, "<=50K" and ">50K". 
# Convert it to a numeric value using StringIndexer.
labelToIndex = StringIndexer(inputCol="Flg", outputCol="label")
```

```
stringIndexerModel = stringIndexer.fit(trainDF)
display(stringIndexerModel.transform(trainDF))
```

```
from pyspark.ml.feature import VectorAssembler
 
# This includes both the numeric columns and the one-hot encoded binary vector columns in our dataset.
numericCols =  [x[0] for x in trainDF.dtypes if x[1]  in ['int','tinyint','float'] and x[0] != 'Flg']
categoricalCols = [x[0] for x in trainDF.dtypes if x[1]  not in ['int','tinyint','float','boolean']]
assemblerInputs = [c + "OHE" for c in categoricalCols] + numericCols
vecAssembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
```

```
print(numericCols)
```

```
from pyspark.ml.classification import LogisticRegression
 
lr = LogisticRegression(featuresCol="features", labelCol="label", regParam=1.0)
```

```
from pyspark.ml import Pipeline
 
# Define the pipeline based on the stages created in previous steps.
pipeline = Pipeline(stages=[stringIndexer, encoder, labelToIndex, vecAssembler, lr])
 
# Define the pipeline model.
pipelineModel = pipeline.fit(trainDF)
 
# Apply the pipeline model to the test dataset.
predDF = pipelineModel.transform(testDF)
```

```
display(predDF.select("features", "label", "prediction", "probability"))
```

```
display(predDF.groupby(["label","prediction"]).count())
```

```
display(pipelineModel.stages[-1], predDF.drop("prediction", "rawPrediction", "probability"), "ROC")
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
 
bcEvaluator = BinaryClassificationEvaluator(metricName="areaUnderROC")
print(f"Area under ROC curve: {bcEvaluator.evaluate(predDF)}")
 
mcEvaluator = MulticlassClassificationEvaluator(metricName="accuracy")
print(f"Accuracy: {mcEvaluator.evaluate(predDF)}")

mcEvaluator = MulticlassClassificationEvaluator(metricName="recallByLabel")
print(f"Recall: {mcEvaluator.evaluate(predDF)}")
```
